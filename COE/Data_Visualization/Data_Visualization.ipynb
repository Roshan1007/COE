```python
xaxis=df['EdLevel']
yaxis=df['YearsCode']
plt.figure(figsize=(25,20))
plt.xlabel("Years of Code", size='12')
plt.ylabel("Education Level",size='12')
plt.bar(xaxis,yaxis, color="blue", width=0.3)
plt.title("Graph for Edu. level and Years of Code")
plt.show()
```

```python
x=df['JobSat'].value_counts()
y=df['JobSat'].unique()
plt.pie(x, labels=y)
plt.legend(loc='lower left')
plt.title('Job Satisfaction')
plt.show() 
```

```python
plt.figure(figsize=(12, 6))
employment_counts = df['EdLevel'].value_counts()
plt.barh(employment_counts.index, employment_counts.values, edgecolor='k', alpha=0.7)
plt.title('Highest Level of Formal Education of Respondents')
plt.xlabel('Number of Respondents')
plt.ylabel('Education Level')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
